## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Drop rows with missing essential fields
netflix.dropna(subset=['title', 'type'], inplace=True)

# Handle 'date_added'
netflix['date_added'] = netflix['date_added'].astype(str).str.strip()
netflix['date_added'] = pd.to_datetime(netflix['date_added'], errors='coerce', format='%B %d, %Y')

# Fill missing 'date_added' with mode (if available)
if not netflix['date_added'].mode().empty:
    most_common_date = netflix['date_added'].mode()[0]
else:
    most_common_date = pd.to_datetime("2000-01-01")
netflix['date_added'] = netflix['date_added'].fillna(most_common_date)

# Create 'year_added' and 'month_added'
netflix['year_added'] = netflix['date_added'].dt.year
netflix['month_added'] = netflix['date_added'].dt.month_name()

# Fill missing 'rating' with mode
if not netflix['rating'].mode().empty:
    rating_mode = netflix['rating'].mode()[0]
    netflix['rating'] = netflix['rating'].fillna(rating_mode)

# Fill missing 'country' and 'director' with 'Unknown'
netflix['country'] = netflix['country'].fillna('Unknown')
netflix['director'] = netflix['director'].fillna('Unknown')

# Clean 'duration'
netflix['duration'] = netflix['duration'].fillna('0 min')
netflix['duration_int'] = netflix['duration'].str.extract('(\d+)').fillna(0).astype(int)
netflix['duration_type'] = netflix['duration'].str.extract('([a-zA-Z]+)').fillna('Unknown')

# Create 'top_country' feature
top_5_countries = netflix['country'].value_counts().head(5).index.tolist()
netflix['top_country'] = netflix['country'].apply(lambda x: 'Top' if x in top_5_countries else 'Other')

# Drop duplicates
netflix.drop_duplicates(inplace=True)
````

### What all manipulations have you done and insights you found?

Missing value treatment ensures accurate visualization and modeling.

Date transformation gives us new temporal insights (year_added, month_added).

Duration split helps differentiate movies (minutes) vs. shows (seasons).

Top country tagging allows for segmentation in analysis.

Dataset is now clean, structured, and analysis-ready.
